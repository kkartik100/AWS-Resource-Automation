```python
import boto3
```

```python
ec2_client = boto3.client("ec2")
ec2_resource = boto3.resource("ec2")
```

```python
#Describe all instances
ec2_client.describe_instances()
```

```python
#Stop particular instances
ec2_client.stop_instances(
    InstanceIds=[
          'i-087d3aa9e3d7c4995',
    ]
)
```

```python
#Start Particular Insatances
ec2_client.start_instances(
    InstanceIds=[
          'i-087d3aa9e3d7c4995',
    ])
```

```python
#Reboot Particular Insatances
ec2_client.reboot_instances(
    InstanceIds=[
          'i-087d3aa9e3d7c4995',
    ])
```

```python
#Terminate Particular Insatances
ec2_client.terminate_instances(
    InstanceIds=[
          'i-087d3aa9e3d7c4995',
    ])
```
